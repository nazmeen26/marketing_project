```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_up=pd.read_csv("digital_marketing_campaign_dataset.csv")
```

```python
df_up.head()
```

```python
df_up.info()
```

```python
df_up.isnull().sum()
```

```python
df_up['Gender'].unique()
```

```python
df_up["Gender"]=np.where(df_up["Gender"].str.contains('Female'),0,1)
```

```python
df_up.head()
```

```python
df_up['Gender']=df_up['Gender'].astype(int)
```

```python
df_up['AdvertisingPlatform'].unique()
```

```python
df_up['AdvertisingTool'].unique()
```

```python
df_up.info()
```

```python
df_up.drop(['AdvertisingTool', 'AdvertisingPlatform'], axis=1, inplace=True)
```

```python
df_up.info()
```

```python
df_up.head()
```

```python
campaigncha=df_up.groupby('CampaignChannel')['Conversion'].mean()
```

```python
campaigncha
```

```python
df_up['Campaign_Channel']=df_up['CampaignChannel'].map(campaigncha)
```

```python
df_up.head(5)
```

```python
df_up.drop('CampaignChannel', axis=1, inplace=True)
```

```python
df_up.info()
```

```python
df_up['CampaignType'].unique()
```

```python
camptype=df_up.groupby('CampaignType')['Conversion'].mean()
```

```python
camptype
```

```python
df_up['Campaign_Type']=df_up['CampaignType'].map(camptype)
```

```python
df_up.head()
```

```python
df_up.drop('CampaignType', axis=1, inplace=True)
```

```python
df_up.info()
```

```python
df_up['Conversion'].value_counts()
```

```python
df_minority=df_up[df_up['Conversion']==0]
df_majority=df_up[df_up['Conversion']==1]
```

```python
from sklearn.utils import resample 
df_minority_upsampled=resample(df_minority, replace=True,
         n_samples=len(df_majority),
         random_state=42
         )
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled=pd.concat([df_majority, df_minority_upsampled])
```

```python
df_upsampled['Conversion'].value_counts()
```

```python
df_upsampled.corr()
```

```python
correlation_matrix = df_upsampled.corr()
plt.figure(figsize=(21, 16))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',fmt=".2f") 
plt.title('Correlation Matrix')
plt.show()
```

```python
df_upsampled.to_csv('digital_marketing_campaign_cleaned_dataset.csv',index=False)
```
